# Import relevant Libraries 

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("BikeStoreData.csv")
data

```

## Check the Data types of each column

```python
data.dtypes
```

## Make Changes to the Date columns (Date Type)

```python
data['order_date'] = pd.to_datetime(data['order_date'], format='%Y-%m-%d')
data['shipped_date'] = pd.to_datetime(data['order_date'], format='%Y-%m-%d')
data['required_date'] = pd.to_datetime(data['order_date'], format='%Y-%m-%d')
data.dtypes
```

## Check for any missing values

```python
data.isnull().sum()
```

```python
data.describe()
```

## Rename the Columns Appropriately

```python
data.rename(columns = {'quantity': 'Order Quantity' , 'first_name' : 'Sales Consultant First Name'}, inplace=True)


data
```

## Add a column Name for total Price based on Quanttity.

```python
# create a new column with the product of "List Price" and "Order Quantity"
data["Total Price"] = data["list_price"] * data["Order Quantity"]

data.head()
```

```python
data.to_csv('CleanBikeStoreData.csv', index=False)
print("Saved")
```
